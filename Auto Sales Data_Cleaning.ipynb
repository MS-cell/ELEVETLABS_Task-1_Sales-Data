```python
import pandas as pd
import numpy as np
# Load Excel file
df = pd.read_excel("Auto Sales data.xlsx")
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert and format order_date
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce', dayfirst=True)
df['order_date'] = df['order_date'].dt.strftime('%d-%m-%Y')

# Standardize text columns
for col in ['status', 'country', 'deal_size']:
    df[col] = df[col].str.strip().str.capitalize()

# Remove duplicates
df = df.drop_duplicates()

# Export cleaned data
df.to_excel("auto_sales_cleaned.xlsx", index=False)
df.to_csv("auto_sales_cleaned.csv", index=False)

print("✅ Data cleaning complete and files exported.")
```
